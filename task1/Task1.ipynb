```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('Iris.csv')
print(f"Original Dataset Shape: {df.shape}")
```


**Visualization:**

*Scatter Plot:*

```python
plt.figure(figsize=(8, 5))
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species', palette='Set2')
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```

*Histogram:*

```python
features = ['SepalLengthCm', 'SepalWidthCm', 'PetalLengthCm', 'PetalWidthCm']

plt.figure(figsize=(8, 5))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.histplot(data=df, x=feature, hue='Species', kde=True, palette='Set2', multiple='stack')
    plt.title(f'Distribution of {feature}',fontsize=10)

plt.tight_layout()
plt.show()
```

*Box Plot:*

```python
plt.figure(figsize=(8, 5))
for i, feature in enumerate(features, 1):
    plt.subplot(2, 2, i)
    sns.boxplot(data=df, x='Species', y=feature, palette='Set2')
    plt.title(f'Box Plot of {feature}',fontsize=10)

plt.tight_layout()
plt.show()
```
